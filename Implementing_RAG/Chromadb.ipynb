```python
import chromadb
```

```python
chroma_client = chromadb.Client()
```

```python
phrases=[
    "Amanda baked cookies and will bring Jerry some tomorrow.",
    "Olivia and Olivier are voting for liberals in this election.",
    "Sam is confused, because he overheard Rick complaining about him as a roommate. Naomi thinks Sam should talk to Rick. Sam is not sure what to do.",
    "John's cookies were only half-baked but he still carries them for Mary."
]
```

```python
ids=["001","002","003","004"] # unique ids for
```

```python
# chromadb expects metadata to be associated with the embeddings
metadatas=[{"source": "pdf-1"},{"source": "doc-1"},{"source": "pdf-2"},{"source": "txt-1"}]
```

```python
metadatas
```

```python
collection = chroma_client.create_collection(name="embeddings_demo") # database which contains multiple documents
```

```python
collection.add(
    documents=phrases,
    metadatas=metadatas,
    ids=ids
)  # downloading our embedded model
```

```python
collection.peek()
```

```python
results = collection.query(
    query_texts=["Mary got half-baked cake from John"],
    n_results=1
)  # doing semantic search, with our query to the existed chromadb embeddings
```

```python
print(results['documents'])
```

```python
results = collection.query(
    query_texts=["cookies"],
    where={"source": "pdf-1"},
    n_results=1
)
```

```python
print(results['documents'])
```

```python
import pandas as pd
import chromadb
from sentence_transformers import SentenceTransformer
import numpy as np
import os
```

```python
df=pd.read_csv('/content/oscars.csv')
df
```

```python
df=df.loc[df['year_ceremony'] == 2023]
df
```

```python
df=df.dropna(subset=['film'])
df.loc[:, 'category'] = df['category'].str.lower()
df
```

```python
df.loc[:, 'text'] = df['name'] + ' got nominated under the category, ' + df['category'] + ', for the film ' + df['film'] + ' to win the award'

df.loc[df['winner'] == False, 'text'] = df['name'] + ' got nominated under the category, ' + df['category'] + ', for the film ' + df['film'] + ' but did not win'
df
```

```python
client = chromadb.Client()

collection = client.get_or_create_collection("oscars-2023")
```

```python
docs=df["text"].tolist()
docs
```

```python
docs[0]
```

```python
# creating unique Ids , to store in chromadb collections.
ids= [str(x) for x in df.index.tolist()]

ids
```

```python
collection.add(
    documents=docs,
    ids=ids
) # adding data in chromadb
```

```python
collection.peek()
```

```python
results = collection.query(
    query_texts=["RRR"],
    n_results=3
)
```
